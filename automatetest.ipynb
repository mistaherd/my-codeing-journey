```python
import re
import tkinter as tk
from tkinter import filedialog
from datetime import datetime, timedelta 
from cryptography.fernet import Fernet

class OutlookWindow:
    def __init__(self):
        self.root = tk.Tk()
        self.root.withdraw()

    def ask_directory(self):
        selected_directory = filedialog.askdirectory()
        return selected_directory
    def exclude_file_extenions(Self):
        file_extensions = [".png",".jpg",".gif",".ARTask"]
        return file_extensions
        
    def mainloop(self):
        self.root.mainloop()

class Outlook:
    def __init__(self, output_directory, api, days_received, not_allowed):
        self.dir = output_directory
        self.api = api
        self.day_receive = days_received
        self.not_allowed = not_allowed

    def check_variables(self):
        if self.api != "Mapi":
            self.api = "Mapi"

        self.day_receive = int(input("Please enter the number of how many days: "))
        self.day_receive = datetime.now()-timedelta(days=self.day_receive)
       
        directory_pattern = r'^[A-Za-z0-9_\-./]+$'

        if re.match(directory_pattern, self.dir):
            outlook_window = OutlookWindow()
            selected_directory = outlook_window.ask_directory()
            self.dir = selected_directory

        if bool(self.not_allowed):
            if bool(int(input("enter 1 : want to exclude not usefull files ,enter 0 : for all files"))):
                    window_1 = OutlookWindow()
                    self.not_allowed = window_1.exclude_file_extenions()
            else:
                self.not_allowed = None

        else:        
            self.not_allowed = None

        attributes = vars(self)  # Get the instance's attributes as a dictionary
        return attributes
    
    def save_attributes(self, filename, key):
        attributes = vars(self)  # Get the instance's attributes as a dictionary
        data = "\n".join([f"{name}: {value}" for name, value in attributes.items()])

        f = Fernet(key)
        encrypted_data = f.encrypt(data.encode())

        with open(filename, "wb") as file:
            file.write(encrypted_data)



# Example usage
obj = Outlook("output", "API key", 1, True)
obj.check_variables()
obj.save_attributes("attributes.txt", b'encryption_key')
```

```python
from docx import Document
import pandas as pd
def open_lastest_word_doc(word_doc_name):
    doc = docx.Document(word_doc_name)
    tables=[]
    for table in doc.tables:
        df =[['' for i in range(len(table.columns))]for j in range(len(table.rows))]
        for i, row in enumerate(table.rows):
            for j, cell in enumerate(row.cells):
                if cell.text:
                    df[i][j] =cell.text
        tables.append(pd.DataFrame(df))
    tables
              
```

```python
import ctypes
import pandas as pd
import keyboard as kb
import time

CF_TEXT = 1
kernel32 = ctypes.windll.kernel32
kernel32.GlobalLock.argtypes = [ctypes.c_void_p]
kernel32.GlobalLock.restype = ctypes.c_void_p
kernel32.GlobalUnlock.argtypes = [ctypes.c_void_p]
user32 = ctypes.windll.user32
user32.GetClipboardData.restype = ctypes.c_void_p

def get_clipboard_text():
    user32.OpenClipboard(0)
    try:
        if user32.IsClipboardFormatAvailable(CF_TEXT):
            data = user32.GetClipboardData(CF_TEXT)
            data_locked = kernel32.GlobalLock(data)
            text = ctypes.c_char_p(data_locked)
            value = text.value
            kernel32.GlobalUnlock(data_locked)
            new_value=value.decode("utf-8")
            print(new_value)
            return new_value
    finally:
        user32.CloseClipboard()

def promptuser():
    import re
    userprompt = input("how many CRQs do you want ? ")
    
    while(not userprompt.isnumeric()):
        userprompt = input("please enter a input that is a number: ")
    userprompt = int(userprompt)
    return userprompt


def main():
    list=[]
    count= 0
    promptnumb =promptuser()-1
    
    
    while(1):
        list.append(get_clipboard_text())
        count = count+1
        promptnumb = promptnumb-1
        if(input("do you want to continue y/n: ")=='n' or promptnumb== 0):
            if(promptnumb != 0 and input("are you sure want to exit your request for {} copies has not been forfilled y/n: ".format(promptnumb))== 'y'):
                break
        else:
            time.sleep(2)
            print("you have printed {} so far out of {} so far".format(count+1,promptnumb))    
            
    clipboard_Session= pd.Series(list)
    data = pd.DataFrame(clipboard_Session)
    data.head()
    print(clipboard_Session)

    data.to_csv("data.csv")

main()
     
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Carrier wave c(t)=A_c*cos(2*pi*f_c*t)
#Modulating wave m(t)=A_m*cos(2*pi*f_m*t)
#Modulated wave s(t)=A_c[1+mu*cos(2*pi*f_m*t)]cos(2*pi*f_c*t)

A_c = float(input('Enter carrier amplitude: '))
f_c = float(input('Enter carrier frquency: '))
A_m = float(input('Enter message amplitude: '))
f_m = float(input('Enter message frquency: '))
modulation_index = float(input('Enter modulation index: '))

t = np.linspace(0, 1, 1000)

carrier = A_c*np.cos(2*np.pi*f_c*t)
modulator = A_m*np.cos(2*np.pi*f_m*t)
product = A_c*(1+modulation_index*np.cos(2*np.pi*f_m*t))*np.cos(2*np.pi*f_c*t)

plt.subplot(3,1,1)
plt.title('Amplitude Modulation')
plt.plot(modulator,'g')
plt.ylabel('Amplitude')
plt.xlabel('Message signal')

plt.subplot(3,1,2)
plt.plot(carrier, 'r')
plt.ylabel('Amplitude')
plt.xlabel('Carrier signal')

plt.subplot(3,1,3)
plt.plot(product, color="purple")
plt.ylabel('Amplitude')
plt.xlabel('AM signal')

plt.subplots_adjust(hspace=1)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(16, 9)
plt.show()
```
